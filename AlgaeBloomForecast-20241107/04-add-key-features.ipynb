```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 
df = pd.read_csv('/root/Download/Modis-algae-aliyun/AlgaeBloomForecast-20241107/merged_all_data_filtered.csv')

# 
target_cols = ['density_mean_y', 'density_lower_y', 'density_upper_y', 
               'chla_mean', 'chla_lower', 'chla_upper']

# ,density_x
exclude_cols = target_cols + ['date', 'density_mean_x', 'density_lower_x', 'density_upper_x']
feature_cols = df.select_dtypes(include=[np.number]).columns.difference(exclude_cols)

# 
corr_matrix = df[list(feature_cols) + target_cols].corr()

# 
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix[target_cols].loc[feature_cols], 
            annot=True, 
            cmap='RdBu_r',
            center=0,
            vmin=-1, 
            vmax=1,
            fmt='.2f')
plt.title('')
plt.tight_layout()
plt.show()

# # 
# for target in target_cols:
#     # 5
#     top_corr = corr_matrix[target].abs().sort_values(ascending=False)[1:6]
#     top_features = top_corr.index
    
#     fig, axes = plt.subplots(2, 3, figsize=(20, 12))
#     fig.suptitle(f'{target}', fontsize=16)
    
#     for i, feature in enumerate(top_features):
#         row = i // 3
#         col = i % 3
#         sns.scatterplot(data=df, x=feature, y=target, ax=axes[row, col])
#         axes[row, col].set_title(f': {corr_matrix.loc[feature, target]:.2f}')
    
#     plt.tight_layout()
#     plt.show()

# 
for target in target_cols:
    print(f"\n{target}:")
    top_corr = corr_matrix[target].abs().sort_values(ascending=False)[1:6]
    for feature, corr in top_corr.items():
        print(f"{feature}: {corr:.3f}")
```

```python
# 
max_lag = 15
lag_corrs = {}

# 
for target in target_cols:
    lag_corrs[target] = {}
    
    # 
    features = feature_cols
    
    # 
    for feature in features:
        lag_corrs[target][feature] = []
        
        # 
        for lag in range(1, max_lag + 1):
            corr = df[feature].shift(lag).corr(df[target])
            lag_corrs[target][feature].append(corr)

# 
for target in target_cols:
    plt.figure(figsize=(15, 8))
    
    # 10
    top_features = corr_matrix[target].abs().sort_values(ascending=False)[1:11].index
    
    for feature in top_features:
        if feature in lag_corrs[target]:
            plt.plot(range(1, max_lag + 1), lag_corrs[target][feature], 
                    marker='o', label=feature)
    
    plt.title(f'{target}')
    plt.xlabel('')
    plt.ylabel('')
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# 
for target in target_cols:
    print(f"\n{target}:")
    
    # 
    feature_lags = {}
    for feature in feature_cols:
        if feature in lag_corrs[target]:
            best_lag = np.argmax(np.abs(lag_corrs[target][feature])) + 1
            best_corr = lag_corrs[target][feature][best_lag-1]
            feature_lags[feature] = (best_lag, best_corr)
    
    # 10
    sorted_features = sorted(feature_lags.items(), key=lambda x: abs(x[1][1]), reverse=True)[:10]
    for feature, (lag, corr) in sorted_features:
        print(f"{feature}: {lag}, {corr:.3f}")
```

```python
# CSV
df = pd.read_csv('merged_all_data_filtered.csv')

# 
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# 
def split_weather(weather):
    if pd.isna(weather):
        return []
    return str(weather).split('-')

# 
all_weather_types = set()
for weather in df['weather'].dropna():
    all_weather_types.update(split_weather(weather))

# 
for weather_type in all_weather_types:
    df[f'weather_{weather_type}'] = df['weather'].apply(lambda x: 1 if pd.notna(x) and weather_type in split_weather(x) else 0)

# 
weather_features = [col for col in df.columns if col.startswith('weather_')]

# 
weather_lag_corrs = {}

# 
for target in target_cols:
    weather_lag_corrs[target] = {}
    
    # 
    for feature in weather_features:
        weather_lag_corrs[target][feature] = []
        
        # 
        for lag in range(1, max_lag + 1):
            # 
            feature_lagged = df[feature].shift(lag)
            target_values = df[target]
            
            # 
            mask1 = feature_lagged.isna()
            mask2 = target_values.isna() 
            valid_mask = ~(mask1 | mask2)
            
            if valid_mask.sum() > 0:
                corr = feature_lagged[valid_mask].astype(float).corr(target_values[valid_mask].astype(float))
            else:
                corr = np.nan
            weather_lag_corrs[target][feature].append(corr)

# 
for target in target_cols:
    plt.figure(figsize=(15, 8))
    
    # 
    weather_corrs = {}
    for feat in weather_features:
        valid_mask = ~(df[feat].isna() | df[target].isna())
        if valid_mask.sum() > 0:
            weather_corrs[feat] = df[feat][valid_mask].astype(float).corr(df[target][valid_mask].astype(float))
        else:
            weather_corrs[feat] = np.nan
    weather_corrs = pd.Series(weather_corrs)
    
    # 10
    top_weather_features = weather_corrs.abs().sort_values(ascending=False)[:10].index
    
    for feature in top_weather_features:
        plt.plot(range(1, max_lag + 1), weather_lag_corrs[target][feature], 
                marker='o', label=feature.replace('weather_',''))
    
    plt.title(f'{target}')
    plt.xlabel('()')
    plt.ylabel('')
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# 
for target in target_cols:
    print(f"\n{target}:")
    
    # 
    weather_feature_lags = {}
    for feature in weather_features:
        if any(not np.isnan(x) for x in weather_lag_corrs[target][feature]):
            best_lag = np.nanargmax(np.abs(weather_lag_corrs[target][feature])) + 1
            best_corr = weather_lag_corrs[target][feature][best_lag-1]
            weather_feature_lags[feature] = (best_lag, best_corr)
    
    # 10
    sorted_weather_features = sorted(weather_feature_lags.items(), key=lambda x: abs(x[1][1]), reverse=True)[:10]
    for feature, (lag, corr) in sorted_weather_features:
        print(f"{feature.replace('weather_','')}: {lag}, {corr:.3f}")
```
